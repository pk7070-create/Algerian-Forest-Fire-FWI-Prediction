```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Algerian_cleaned.csv")
df
```

```python
df.drop(['month','day','year'],axis=1,inplace=True)
```

```python
#encoding
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
df
```

```python
#MAKING AN INDEPENDENT AND DEPENDENT FEATURES.
X=df.drop('FWI',axis=1)
y=df['FWI']
X
```

```python
y
```

```python
#train and test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.25,random_state=42
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.corr()
```

```python
#checking for multicollinearity.
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)

```

```python
#dropping the features when correlation is more than 0.85

def correlation(dataset, threshold):

    col_corr = set()

    corr_matrix = dataset.corr()

    for i in range(len(corr_matrix.columns)):

        for j in range(i):

            if abs(corr_matrix.iloc[i, j]) > threshold:

                colname = corr_matrix.columns[i]

                col_corr.add(colname)

    return col_corr
```

```python
cor_features=correlation(X_train,0.85)
cor_features
```

```python
print(X_train.columns.tolist())
```

```python
X_train.drop(cor_features, axis=1, inplace=True, errors='ignore')
X_test.drop(cor_features, axis=1, inplace=True, errors='ignore')
X_train.shape,X_test.shape
```

FEATURE SCALING OR STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
X_train_scaled=scalar.fit_transform(X_train)
X_test_scaled=scalar.transform(X_test)
```

```python
X_train_scaled,X_test_scaled
```

BOX PLOT TO UNDERSTAND THE EFFECT OF STANDARDIZATION 

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train before scaling")

plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train after scaled")
```

LINEAR REGRESSION MODEL

```python
print(X_train_scaled.shape)
print(X_test_scaled.shape)

print(X_train.columns)
print(X_test.columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error ,r2_score
linreg=LinearRegression(n_jobs=-1)
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
R2_score=r2_score(y_test,y_pred)
print(mae)
print(R2_score)
plt.scatter(y_test,y_pred)
plt.title("Linear regression model")
```

LASSO REGRESSION

```python
#checking with LASSO
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error ,r2_score
Lasso=Lasso()
Lasso.fit(X_train_scaled,y_train)
y_pred=Lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
R2_score=r2_score(y_test,y_pred)
print(mae)
print(R2_score)
plt.scatter(y_test,y_pred)
plt.title("Lasso_regression")
```

```python
#cross validation on LASSO
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
lassocv.alpha_
```

```python
lassocv.alphas_
```

```python
lassocv.mse_path_
```

RIDGE REGRESSION

```python
#checking with LASSO
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error ,r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
R2_score=r2_score(y_test,y_pred)
print(mae)
print(R2_score)
plt.scatter(y_test,y_pred)
plt.title("Ridge_regression")
```

ELASTIC NET REGRESSION

```python
#checking with LASSO
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error ,r2_score
ek=ElasticNet()
ek.fit(X_train_scaled,y_train)
y_pred=ek.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
R2_score=r2_score(y_test,y_pred)
print(mae)
print(R2_score)
plt.scatter(y_test,y_pred)
plt.title("Elasticnet_regression")
```

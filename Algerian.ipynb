DATA CLEANING PROCESS

```python
#Finding the missing values.
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
#making the new column and dividing that group 
df.loc[:122 , "Region"] = 0
df.loc[122: , "Region"] = 1
df.head()
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#dropping the missing values
df=df.dropna().reset_index(drop=True)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.iloc[122:]
```

```python
#dropping the 122nd row
df=df.drop(122).reset_index(drop=True)
df
df.iloc[122:]
```

```python
#fixing the extra space in the column names
df.columns=df.columns.str.strip()
df.columns
```

```python
#changing the required columns to int type
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
#converting the required columns into float
df[['Rain','FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain','FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
df.info()
```

```python
#again saving the cleaned dataset into a new csv file.
df.to_csv("Algerian_cleaned.csv", index=False)
```

EXPLORATORY DATA ANALYSIS.

```python
df_copy= df.drop(['day','month','year'], axis=1)
df_copy
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
df_copy
```

```python
df_copy['Classes'].value_counts()
```

```python
#some kind of vizualization
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8')

df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
##Percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
percentage
```

```python
#pie chart 
classlabels=['Fire','Not Fire']
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Percentage of Fire and Not Fire")
plt.show()
```

```python
#correlation
df_copy.corr()
```

```python
#heatmap for correlation
sns.heatmap(df_copy.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
